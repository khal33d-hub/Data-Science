## Model Fitting & Evaluation
Predicting whether a recipe is high traffic or Not high traffic is a **classification problem in machine learning**. 

I am choosing logistic regression model as the base model this is a simple model to implement but it can not interprete the non-linear relationship in our data. However, it would serve as a performance benchmark against a more complex comparision model. 

The comparision model is Random Forest Classifier, it is a more complex model that can capture non-linear relationship in the dataset and it can be used for feature importance to understand which features are most influential.

For evaluation, i am choosing the accuracy and precision percentage to evaluate both models. Accuracy is the overall percentage of correct predictions, helping to assess the model's general performance. Precision is a measure of how many of the correct predictions made by the model were actually correct, this is crucial as we want to make sure the website is always showing high traffic food no matter what.

### Prepare Data for Modelling
To enable modelling, I chose _calories,sugar,protein,carbohydrate, servings and category_ as **features** and _high_traffic_ as **target variable**. I also have made the following changes: 
- I dropped the recipe column because it contains unique values not relevant to the prediction
- Standardised the numeric features.
- Encode the categorical features into numeric data.
- Convert the target variable into binary labels.
- Adjust the train/test split for class imbalance in target variable.
- Split the data into a training set and a test set.

```python
# Encoding categorical variables
categorical_features = ['category']
numeric_features = ['calories', 'carbohydrate', 'sugar', 'protein', 'servings']

# Convert 'high_traffic' to binary labels
df['high_traffic'] = df['high_traffic'].apply(lambda x: 1 if x == 'High' else 0)
```

```python
# transforming the numeric and categorical features
# Create a column transformer for preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Feature Selection (dropping target variable and recipe)
# Splitting the dataset into the Training set and Test set, adjusting for class imbalance in target
X = df.drop(['high_traffic', 'recipe'], axis=1)
y = df['high_traffic']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2011)
```

### Logistic Regression Model

```python
# Creating a pipeline with preprocessing and model
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('classifier', LogisticRegression())])

# Train the model
model.fit(X_train, y_train)

# Performance Evaluation
# Predicting the Test set results
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Calculating the metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)

print(f'Accuracy: {accuracy * 100:.2f}%')
print(f'Precision: {precision * 100:.2f}%')
```

#### Finding Feature Importance
The logistic Regression model shows us its top features that determine high traffic are _vegetable_, _potatoes_ and _pork_. This would not be investigated further because the model cannot capture the non-linear relationship in our data.

```python
# Model Interpretation - Feature Importance
importance = model.named_steps['classifier'].coef_[0]


# Get feature names after one-hot encoding
feature_names = model.named_steps['preprocessor'].named_transformers_['cat'].get_feature_names_out()

# Combine numeric and categorical feature names
all_feature_names = np.concatenate([numeric_features, feature_names])

# Combine names and importances and sort by importance
combined = zip(all_feature_names, importance)
sorted_features = sorted(combined, key=lambda x: x[1], reverse=True)  # Sort by importance

# Unzip the sorted tuples
sorted_feature_names, sorted_importance = zip(*sorted_features)

# Create a DataFrame for Seaborn
importance_df = pd.DataFrame({'Feature': sorted_feature_names, 'Importance': sorted_importance})

# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df, orient='h')
plt.title('Feature Importance in Logistic Regression Model')
# plt.gca().invert_yaxis()  # Optional: Invert y-axis to have the largest bar at the top
plt.show()
```

### Decision Tree Classifier Model

```python
# Creating a pipeline with preprocessing and model
rf_model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('classifier', RandomForestClassifier(min_samples_split=10, min_samples_leaf=2, random_state=2011))])

# Training the Random Forest model
rf_model.fit(X_train, y_train)

# Predicting the test set results
y_pred_2 = rf_model.predict(X_test)
y_pred_proba_2 = rf_model.predict_proba(X_test)[:, 1]

# Calculating performance metrics
accuracy_rf = accuracy_score(y_test, y_pred_2)
precision_rf = precision_score(y_test, y_pred_2)

print(f'Accuracy: {accuracy_rf * 100:.2f}%')
print(f'Precision: {precision_rf * 100:.2f}%')
```

#### Hyperparameter Tuning

```python
# Define a range of hyperparameters for tuning
param_grid = {
    'classifier__max_depth': [None, 10, 15, 20],  # Maximum depth of the tree
    'classifier__min_samples_split': [2, 5, 10],  # Minimum number of samples required to split a node
    'classifier__min_samples_leaf': [1, 2, 4]     # Minimum number of samples required at each leaf node
}

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, 
                           cv=3, n_jobs=-1, verbose=2, scoring='precision')

# Perform grid search
grid_search.fit(X_train, y_train)

# Best hyperparameters
print("Best hyperparameters:\n", grid_search.best_params_)
```

#### Finding Feature Importance
The Random Forest Classifier model shows us its top features that determine high traffic are _protein_, _Beverages_, _sugar_ and _calories_. This is the chosen feature importance because the model can interprete the non-linear relationship we have in our data.

```python
# Extracting feature importances
feature_importances = rf_model.named_steps['classifier'].feature_importances_

# Handling feature names for one-hot encoded categories
ohe_categories = rf_model.named_steps['preprocessor'].named_transformers_['cat'].categories_
new_feature_names = [f'{col}_{cat}' for col, cats in zip(categorical_features, ohe_categories) for cat in cats]
all_feature_names = numeric_features + new_feature_names

# Mapping importances to features
importance_dict = dict(zip(all_feature_names, feature_importances))

# Sorting features by importance
importances_sorted = sorted(importance_dict.items(), key=lambda x: x[1], reverse=True)

# Creating a DataFrame for easier plotting
importances_df = pd.DataFrame(importances_sorted, columns=['Feature', 'Importance'])

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importances_df)
plt.title('Feature Importances in Random Forest Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

### Evaluate by Business Criteria
To align with the product team's goal of minimizing the promotion of low-traffic recipes and accurately predicting high-traffic recipes at least 80% of the time, we have re-evaluated our models with an adjusted prediction probability threshold. By increasing the threshold from 50% to 80%, we aim to focus more stringently on high-traffic predictions. This adjustment serves as a key performance indicator (KPI), enabling us to reassess the efficacy of both the Logistic Regression and Random Forest models under these new criteria.

**Upon this re-evaluation, the precision of the Logistic Regression model stands at 96.08%, while the Random Forest model exhibits a precision of 95.56%.** The negligible difference of less than 1% between the two models indicates comparable performance under the heightened threshold. This further reinforces that the Random Forest Classifier maintains robust performance in comparison to the Logistic Regression model.

Despite the Logistic Regression model's marginally higher precision, the difference is less than 1%. **Consequently, our recommendation remains in favor of the Random Forest model, particularly due to its demonstrated capability to effectively interpret the non-linear relationships present in our dataset.**

```python
# Setting a threshold for classification
threshold = 0.8  # 80% probability to predict 'high' traffic
y_pred_thresholded = (y_pred_proba >= threshold).astype(int)

# Performance Evaluation with adjusted threshold
accuracy_thresholded = accuracy_score(y_test, y_pred_thresholded)
precision_thresholded = precision_score(y_test, y_pred_thresholded)

print(f"Logistic Regression with Threshold {threshold * 100}%:")
print(f"Accuracy is: {accuracy_thresholded * 100:.2f}%")
print(f"Precision is: {precision_thresholded * 100:.2f}%")
```

```python
# Logistic regression evaluation precision metrics at different thresholds
metrics = {
    'Threshold': ['Original (50%)', 'Adjusted (80%)'],
    'Precision': [precision, precision_thresholded]
}

# Convert the data to a DataFrame
metrics_df = pd.DataFrame(metrics)

# Melting the DataFrame for easier plotting with Seaborn
metrics_df_melted = pd.melt(metrics_df, id_vars='Threshold', value_vars=['Precision'])

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Threshold', y='value', data=metrics_df_melted)
plt.title('Logistic Regression Performance at Different Thresholds')
plt.ylabel('Score')

# Draw a red dotted line at the height of the first bar's precision value
plt.axhline(y=precision, color='red', linestyle='--')

plt.show()
```

```python
# Setting a new threshold for classification
threshold = 0.8  # Adjust this threshold to achieve 80% precision
y_pred_adjusted = np.where(y_pred_proba_2 >= threshold, 1, 0)

# Evaluating model performance with the adjusted threshold
accuracy_adjusted = accuracy_score(y_test, y_pred_adjusted)
precision_adjusted = precision_score(y_test, y_pred_adjusted)

print(f"Random forest Classifier with Adjusted Threshold ({threshold * 100}%):")
print(f"Accuracy is: {accuracy_adjusted * 100:.2f}%")
print(f"New Precision is: {precision_adjusted * 100:.2f}%")
```

```python
# Random Forest Classifier evaluation precision metrics at different thresholds
metrics = {
    'Threshold': ['Original (50%)', 'Adjusted (80%)'],
    'Precision': [precision_rf, precision_adjusted]
}

# Convert the data to a DataFrame
metrics_df = pd.DataFrame(metrics)

# Melting the DataFrame for easier plotting with Seaborn
metrics_df_melted = pd.melt(metrics_df, id_vars='Threshold', value_vars=['Precision'])

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Threshold', y='value', data=metrics_df_melted)
plt.title('Random Forest Classifier Performance at Different Thresholds')
plt.ylabel('Score')

# Draw a red dotted line at the height of the first bar's precision value
plt.axhline(y=precision_rf, color='red', linestyle='--')

plt.show()
```
